```python
import pandas as pd
import folium
```

```python
data = {
    'City': ['Paris', 'Bangkok', 'London', 'Dubai', 'Singapore', 'New York', 'Istanbul'],
    'Latitude': [48.8566, 13.7563, 51.5072, 25.2769, 1.3521, 40.7128, 41.0082],
    'Longitude': [2.3522, 100.5018, -0.1276, 55.2962, 103.8198, -74.0060, 28.9784],
    'Visitors (millions)': [19.1, 22.78, 19.6, 16.3, 14.67, 13.6, 13.4]
}
df = pd.DataFrame(data)
```

```python
# Create a base map centered on the equator
m = folium.Map(location=[20, 0], zoom_start=2, tiles='OpenStreetMap')

# Add circular markers for each city
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['Visitors (millions)'] * 0.5,  # Scale radius for visibility
        popup=f"{row['City']}: {row['Visitors (millions)']}M visitors",
        color='#3186cc',
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

# Display the map
m
```
